```python
import requests
import time
import matplotlib.pyplot as plt
```

```python
iss_data=requests.get("https://api.wheretheiss.at/v1/satellites/25544")
iss_format=iss_data.json()
print (iss_format['latitude'],iss_format['longitude'], iss_format['timestamp'])
```

```python
latitudes=[]
longitudes=[]
timestamps=[]

for i in range(100):
    iss_data=requests.get("https://api.wheretheiss.at/v1/satellites/25544")
    iss_format=iss_data.json()
    latitude=iss_format['latitude']
    longitude=iss_format['longitude']
    timestamp=iss_format['timestamp']
    latitudes.append(latitude)
    longitudes.append(longitude)
    timestamps.append(timestamp)
    time.sleep(10)
```

```python
fig, axs=plt.subplots(1,3)

axs[0].plot(latitudes, longitudes)
axs[0].set_title("Latitudes vs Longitudes")
axs[0].set_xlabel("Latitude")
axs[0].set_ylabel("Longitude")
axs[0].grid(True)

axs[1].plot(timestamps,latitudes)
axs[1].set_title("Latitude of the ISS")
axs[1].set_xlabel("time")
axs[1].set_ylabel("Latitude")
axs[1].grid(True)

axs[2].plot(timestamps,longitudes)
axs[2].set_title("Longitude of the ISS")
axs[2].set_xlabel("time")
axs[2].set_ylabel("Longitude")
axs[2].grid(True)


plt.tight_layout()
plt.show()
```
